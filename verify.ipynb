```python
from pyspark.sql import SparkSession



spark = SparkSession.builder.appName('Scripts Verify').master("local[*]").getOrCreate()
```

```python
df = spark.read.parquet('data/raw/repos')
```

```python
print("Row count: ", df.count())
print("Schema: \n")
df.printSchema()

df.show()

```

```python
df.collect()
print({row['full_name']: row['repo_id'] for row in df.collect()})
```

```python
spark.stop()
```
